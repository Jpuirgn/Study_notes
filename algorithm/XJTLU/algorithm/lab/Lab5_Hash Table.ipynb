## Part 1: Python Built-in `hash()` Function

A hash function converts an input object (key) into an integer value called a **hash value**.

In Python, we can use the built-in function `hash()` to compute this value.

### Key Points

- `hash(obj)` returns an integer
- Hash values are used in hash tables (e.g., dictionaries)
- Only **hashable (immutable)** objects can be hashed:
  - Examples: `int`, `float`, `str`, `tuple`
- Mutable objects like `list` and `dict` are **not hashable** (Only objects with a **fixed, unchanging value** can be safely used as keys in hash tables.)

### Example

The following code demonstrates how `hash()` works on different data types.

```python
# Examples of hash() usage

print("Integer:", hash(10))
print("String:", hash("hello"))
print("Tuple:", hash((1, 2, 3)))

# Try an unhashable object
try:
    print(hash([1, 2, 3]))
except TypeError as e:
    print("Error:", e)
```

## Part 3: Hash Table in Python (`dict`)

In Python, we usually do not implement hash table manually.  
Instead, we use the built-in **dictionary (`dict`)**, which provides the core functionality of a hash table, including fast key-value lookups.

### Key Ideas

- A dictionary stores **key-value pairs**
- Keys are hashed using `hash()`
- The hash value determines where the data is stored
- This allows **fast insertion, search, and deletion (average O(1))**

### Example

The following example shows how a dictionary behaves like a hash table.

```python
# Using Python dictionary as a hash table

# Create a dictionary
ht = {}

# Insert key-value pairs
ht["apple"] = 3
ht["banana"] = 2
ht["cherry"] = 5

# Search (lookup)
print(ht["banana"])   # 2

# Update value
ht["banana"] = 4
print(ht["banana"])   # 4

# Check existence
print("apple" in ht)   # True
print("durian" in ht)  # False

# Delete a key
del ht["apple"]

print(ht)
```
