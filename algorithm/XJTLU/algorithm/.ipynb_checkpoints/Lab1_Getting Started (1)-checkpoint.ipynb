<h2> Sorting Algoritms </h2>

Sorting refers to arranging data in a particular format. Sorting algorithm specifies the way to arrange data in a particular order. Most common orders are in numerical or lexicographical order.

The importance of sorting lies in the fact that data searching can be optimized to a very high level, if data is stored in a sorted manner. Sorting is also used to represent data in more readable formats. 

In Lab1, we will cover 3 such implementations of sorting algorithms in python.

<ul>
    <li>Insertion Sort</li>
    <li>Selection Sort</li>
    <li>Bubble Sort</li>
</ul>

<h2> Insertion Sort </h2>

Insertion sort involves finding the right place for a given element in a sorted list. So in beginning we compare the first two elements and sort them by comparing them. Then we pick the third element and find its proper position among the previous two sorted elements. This way we gradually go on adding more elements to the already sorted list by putting them in their proper position.

Follow the pseudocode given in the lecture notes to finish the function below.

```python
def insertion_sort(InputList):
    for j in range(1, len(InputList)):
        key = InputList[j]
        i = j - 1
        while i >= 0 and InputList[i] > key:
            InputList[i + 1] = InputList[i]
            i = i - 1
        InputList[i + 1] = key
            
    return InputList

arr = [19,2,31,45,30,11,121,27]
print(insertion_sort(arr))
```

When the above code is executed, it produces the sorted array: [2, 11, 19, 27, 30, 31, 45, 121].

<h2> Selection Sort </h2>

Selection Sort is a simple sorting algorithm that repeatedly selects the smallest (or largest) element from the unsorted portion of the list and moves it to the sorted portion. 

Follow the pseudocode given in the lecture notes to finish the function below.

```python
def selection_sort(array):
    for i in range(0, len(array)):
        min = i
        for j in range(i+1, len(array)):
            if array[j] < array[min]:
                min = j
        array[i], array[min] = array[min], array[i]
    return array

# Example usage
array = [64, 34, 25, 12, 22, 11, 90]
print(selection_sort(array))
```

When the above code is executed, it should produce the sorted array: [11, 12, 22, 25, 34, 64, 90]

<h2> Bubble Sort </h2>

Steps to implement bubble sort:

1. In first cycle,
    1. Start by comparing 1st and 2nd element and swap if 1st element is greater.
    2. After that do the same for 2nd and 3rd element.
    3. At the end of cycle you will get max element at the end of list.
2. Now do the same in all subsequent cycles.
3. Perform this for (number of elements – 1) times.
4. You will get sorted list.

Step-by-step example:

Take an array of numbers "5 1 4 2 8", and sort the array from lowest number to greatest number using bubble sort. In each step, elements written in bold are being compared. 

First Pass\
( **5 1** 4 2 8 ) → ( **1 5** 4 2 8 ), Here, algorithm compares the first two elements, and swaps since 5 > 1.\
( 1 **5 4** 2 8 ) → ( 1 **4 5** 2 8 ), Swap since 5 > 4\
( 1 4 **5 2** 8 ) → ( 1 4 **2 5** 8 ), Swap since 5 > 2\
( 1 4 2 **5 8** ) → ( 1 4 2 **5 8** ), Now, since these elements are already in order (8 > 5), algorithm does not swap them.

Second Pass\
( **1 4** 2 5 8 ) → ( **1 4** 2 5 8 )\
( 1 **4 2** 5 8 ) → ( 1 **2 4** 5 8 ), Swap since 4 > 2\
( 1 2 **4 5** 8 ) → ( 1 2 **4 5** 8 )\
( 1 2 4 **5 8** ) → ( 1 2 4 **5 8** )

Now, the array is already sorted, but the algorithm does not know if it is completed. The algorithm needs additional whole passes without any swap to know it is sorted.

```python
def bubble_sort(array):
    
    # loop to access each array element
    for i in range(0, len(array)):
        

        # loop to compare array elements
        for j in range(0, len(array) - 1):

            # compare two adjacent elements
            # change > to < to sort in descending order
            if array[j] > array[j + 1]:
                array[j], array[j + 1] = array[j + 1], array[j]

    return array


data = [-21, 5, 0, 15, -9, 11, 27, 10]
print(bubble_sort(data))
```

When the above code is executed, it should produce the sorted array: [-21, -9, 0, 5, 10, 11, 15, 27]

<h2> Generating random arrays of integers </h2>

In this section, you will learn how to generate random integer arrays and use them to test different sorting algorithms. 

Hints: Python's random module helps generate random numbers, i.e., random.randint(). Check https://www.w3schools.com/python/ref_random_randint.asp 

(You can also use numpy, https://numpy.org/doc/stable/reference/random/generated/numpy.random.randint.html)


```python
import random
import time

def generate_random_array(size, min_val=0, max_val=1000):
    #Generates a random array of integers
    return [random.randint(min_val, max_val) for _ in range(size)]

# generate a random array of integers, size = 10
random_array = generate_random_array(10)
print("Random array:", random_array)

# Get the sorted array via selection sort algorithm
sorted_selection = selection_sort(random_array.copy())
# Get the sorted array via insertion sort algorithm
sorted_insertion = insertion_sort(random_array.copy())
# Get the sorted array via bubble sort algorithm
sorted_bubble = bubble_sort(random_array.copy())

print("Sorted with Selection Sort:", sorted_selection)
print("Sorted with Insertion Sort:", sorted_insertion)
print("Sorted with Bubble Sort:", sorted_bubble)

# test the sorting algortihms under different input size
for size in [10, 100, 1000, 10000]:
    arr = generate_random_array(size)

    print("\nInput size:", size)
    start = time.time()
    selection_sort(arr.copy())
    end = time.time()
    print("Selection Sort times:", end - start)

    start = time.time()
    insertion_sort(arr.copy())
    end = time.time()
    print("Insertion Sort time:", end - start)

    start = time.time()
    bubble_sort(arr.copy())
    end = time.time()
    print("Bubble Sort time:", end - start)
```

### Questions

1. Insertion sort, selection sort, and bubble sort all have a time complexity of **O(n²)**.  
   Compare their measured running times and explain why their practical performance differs despite having the same asymptotic complexity.

2. For each of the three algorithms, identify the **best-case** and **worst-case** input scenarios and briefly explain why these cases lead to the fastest or slowest performance.

3. Measure the running time of each algorithm on **sorted** and **reverse-sorted** arrays.

Solution:
1. Although insertion sort, selection sort, and bubble sort all have O(n²) worst-case time complexity, their measured running times can be different because Big-O notation only describes the growth rate and ignores constant factors and implementation details.
Insertion sort is often faster on nearly sorted data because it only shifts elements when necessary. Selection sort always scans the remaining unsorted part to find the minimum element, so its running time changes less with input order. Bubble sort may be slower because it repeatedly compares adjacent elements and performs many swaps. Therefore, even with the same asymptotic complexity, the actual running times can differ in practice.

2. For insertion sort, the best case is an already sorted array. In this case, each element only needs one comparison and no major shifting is needed, so the time complexity is O(n). The worst case is a reverse-sorted array because each new element has to be compared with and shifted past many previous elements, giving O(n²).
For selection sort, the best case and worst case are both O(n²). Even if the array is already sorted, the algorithm still scans the entire unsorted part in each pass to find the minimum element. Therefore, the input order does not significantly reduce the number of comparisons.
For bubble sort, the worst case is a reverse-sorted array because many adjacent swaps are needed. If an optimized version with a swapped flag is used, the best case is an already sorted array, because the algorithm can stop after one pass with no swaps, giving O(n). Without this optimization, bubble sort still performs repeated passes and remains O(n²).
